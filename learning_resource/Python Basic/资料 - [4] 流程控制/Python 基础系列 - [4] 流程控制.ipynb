### <font color='#2b4750' face='微软雅黑'>2.1 `while` 循环</font><a name='2.1'></a>
[<font color='black' face='微软雅黑'>回到章首</font>](#2)
```

    <statement(s)>
```

````python
n = 5
while n > 0:
    print(n)
    n = n - 1
print('I love Python')
````

`while` 后面的语句最终返回一个布尔值 (**`True`**, **`False`**)，我们知道非布尔型变量也有相对性的布尔值，设该变量为 X

- 当 X 是元素型变量（整型、浮点型）
    - 当 X 为 0, 0.0, `bool(X) =` **`False`**
    - 当 X 非零, `bool(X) =` **`True`**


- 当 X 是容器型变量（字符串、列表、元组、字典、集合）
    - 当 X 为空, `bool(X) =` **`False`**
    - 当 X 不为空, `bool(X) =` **`True`**

验证如下：

````python
print( type(0), bool(0), bool(1) )
print( type(10.31), bool(0.00), bool(10.31) )
print( type(''), bool( '' ), bool( 'python' ) )
print( type(()), bool( () ), bool( (10,) ) )
print( type([]), bool( [] ), bool( [1,2] ) )
print( type({}), bool( {} ), bool( {'a':1, 'b':2} ) )
print( type(set()), bool( set() ), bool( {1,2} ) )
````

````python
asset_class = ['FX', 'CM', 'EQ']
while asset_class:
    print(asset_class.pop(-1))
````

**枚举函数：`enumerate`**
```
for (idx, var) in enumerate(iterable):
	<statement(s)>
```
如果我们想把每种语言附加对应的索引也打印出来，怎么办？先看看一种**丑陋**的写法。

````python
index = 0
for lang in languages:
    print(index, lang)
    index +=1
````

`enumerate()` 函数可以一次性返回列表的元素以及其对应的索引。

````python
for index, lang in enumerate(languages):
    print(index, lang)
````

我们还可以自定义索引的初始值。在实际生活中，一般索引从 1 开始更自然，那么将参数 `start` 设置为 1 即可。

````python
for index, lang in enumerate(languages, start=1):
    print(index, lang)
````

**打包函数：`zip`**
```
for (i1, i2) in zip(iter1, iter2):
	<statement(s)>
```

给定一串名字（names）和演员角色（actors），用两个列表存储。

````python
names = ['小罗伯特唐尼', '托比·马奎尔', '克里斯蒂安·贝尔', '杰森·莫玛']
actors = ['钢铁侠', '蜘蛛侠', '蝙蝠侠', '水行侠']
````

如果我们想把每个名字和角色一一对应起来，可以用上节学到的 `enumerate()` 函数。我们可以返回 `names` 里的元素和索引，再用索引来获取 `actors` 里的元素。

````python
for index, name in enumerate(names):
    print(f'{name}是{actors[index]}')
````

结果不够优雅，来看看 `zip()` 怎么优雅操作。

````python
for name, actor in zip(names, actors):
    print(f'{name}是{actor}')
````

`zip()` 函数将列表（迭代器）中对应的元素打包成一个个元组，然后返回由这些元组组成的列表。

再加一个列表如何，`zip()` 函数也能轻松处理。

````python
universes = ['漫威', '漫威', 'DC', 'DC']

for name, actor, universe in zip(names, actors, universes):
    print(f'{name}是来自{universe}的{actor}')
````

### <font color='#2b4750' face='微软雅黑'>2.3 `break, continue, else`</font><a name='2.3'></a>
[<font color='black' face='微软雅黑'>回到章首</font>](#2)

`break`, `continue` 和 `else` 对 **`while` 循环**和 **`for` 循环**都适用。

- `break` 终止**整个循环**，并继续执行**循环后的下一条语句**。
- `continue` 终止**当前迭代**，并继续进行**下一轮迭代**。
- `else` 子句仅在**循环正常迭代**终止后才执行，可以理解成 `nobreak` 在条件下。

````python
products = ['European Swap', 'European Option', 'Asian Swap', 'Asian Option', 'Barrier Option']

for product in products:
    if 'Option' in product:
        break
    print(product)
````

````python
for product in products:
    if 'Option' in product:
        continue
    print(product)
````

````python
for product in products:
    if 'Option' in product:
        break
    print(product)
else:
    print('Task done!')
````

````python
for product in products:
    if 'Option' in product:
        continue
    print(product)
else:
    print('Task done!')
````

我们用**面向对象编程**的方式来编写欧式期权在 Black-Scholes 模型下的解析解。首先引进 `numpy` 和 `scipy` 包下的函数。

````python
from numpy import log, sqrt, exp
from scipy import stats
````

````python
class BSM:
    
    def __init__(self, flag, S0, K, r, q, T, sig):
        self.flag = flag.lower()[0]
        self.S0 = S0
        self.K = K;
        self.r = r;
        self.q = q;
        self.T = T
        self.sig = sig
        self.omega = 1 if self.flag=='c' else -1 if self.flag=='p' else np.nan
    
    def d1(self):
        return (log(self.S0/self.K) + ((self.r-self.q)+self.sig**2/2)*self.T) \
                / (self.sig*sqrt(self.T))

    def d2(self):
        return (log(self.S0/self.K) + ((self.r-self.q)-self.sig**2/2)*self.T) \
                / (self.sig*sqrt(self.T))

    def value(self):
        return self.omega * (self.S0*exp(-self.q*self.T)*stats.norm.cdf(self.omega*self.d1()) \
                             - self.K*exp(-self.r*self.T)*stats.norm.cdf(self.omega*self.d2()))
````

<font color='blue'>**Raise 个性化 Exception**</font>

````python
def portfolio_value(last_worth, current_worth):
    if last_worth < 0 or current_worth < 0:
        raise ValueError('Negative worth!')
    value = current_worth - last_worth
    if value < 0:
        raise ValueError('Negative return!')
````

````python
try:
    portfolio_value(-10000, 10001)
except ValueError as err:
    print(err)
````

````python
try:
    portfolio_value(10000, 9999)
except ValueError as err:
    print(err)
````

````python
class Error(Exception):
    pass

class NegativePortfolioValueWarning(Error):
    def __init__(self, expression, message):
        self.expression = expression
        self.message = message
````

````python
def portfolio_value(last_worth, current_worth):
    if last_worth < 0 or current_worth < 0:
        raise ValueError('Negative worth!')
    value = current_worth - last_worth
    if value < 0:
        raise NegativePortfolioValueWarning('NegativePortfolioValueWarning', \
                                            'Negative return. Take a look!')
````

````python
try:
    portfolio_value(-10000, 10001)
except ValueError as err:
    print(err)
except NegativePortfolioValueWarning as err:
    print('[', err.args[0], '] -', err.args[1])
````

````python
try:
    portfolio_value(10000, 9999)
except ValueError as err:
    print(err)
except NegativePortfolioValueWarning as err:
    print('[', err.args[0], '] -', err.args[1])
````

### <font color='#2b4750' face='微软雅黑'>4.4 `try-except-else-finally` 范式</font><a name='4.4'></a>
[<font color='black' face='微软雅黑'>回到章首</font>](#4)

````python
def divide(a, b):
    try:
        c = a / b
    except ZeroDivisionError:
        print('Divisor is zero and division is impossible!')
    else:
        print(f"Result = {c:.4f}.")
    finally:
        print('Error or no error, FINALLY DONE!')
````

````python
divide(10, 0)
````

````python
divide(10, 2)
````

````python
def divide(a, b):
    try:
        c = a / b
    except ZeroDivisionError:
        print('Divisor is zero and division is impossible!')
    except Exception:
        print('Something wrong!')
    else:
        print(f"Result = {c:.4f}.")
    finally:
        print('Error or no error, FINALLY DONE!')
````

````python
divide(10, '2')
````
